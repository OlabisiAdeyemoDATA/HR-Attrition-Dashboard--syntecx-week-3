# HR Analytics Correlation Analysis

This notebook explores factors influencing employee attrition using the Atlas Labs HR dataset.

Objectives:
- Identify factors associated with employee attrition.
- Analyze relationships between salary, age and experience.
- Discover retention patterns.
- Generate insights to support HR decision-making.


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('employee_cleaned.csv')

df.head()
```

```
df.info()
```

```
df['Attrition_Flag'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

## Data Preparation

The employee dataset was loaded and cleaned before analysis.

An Attrition_Flag variable was created:

- Yes = 1
- No = 0

This enables numerical correlation analysis.

```
correlation_columns = [
    'Age',
    'Salary',
    'DistanceFromHome (KM)',
    'YearsAtCompany',
    'YearsInMostRecentRole',
    'YearsSinceLastPromotion',
    'YearsWithCurrManager',
    'Attrition_Flag'
]

corr_matrix = df[correlation_columns].corr()

corr_matrix
```

## Correlation Analysis

Correlation analysis was performed on:

- Age
- Salary
- Distance from Home
- Years at Company
- Years in Most Recent Role
- Years Since Last Promotion
- Years With Current Manager

These variables were compared against employee attrition.

```
attrition_corr = corr_matrix['Attrition_Flag'].sort_values(ascending=False)

attrition_corr
```

## Key Findings

1. Employees who have spent fewer years at the company are more likely to leave.

2. Employees with fewer years since their last promotion show higher attrition.

3. Employees with shorter relationships with their current manager are more likely to leave.

4. Salary has a weak negative relationship with attrition.

5. Distance from home has almost no effect on attrition.

```
# Total Employees
total_employees = len(df)

# Attrition Rate
attrition_rate = (df['Attrition'] == 'Yes').mean() * 100

# Retention Rate
retention_rate = 100 - attrition_rate

# Average Salary
avg_salary = df['Salary'].mean()

# Average Experience
avg_experience = df['YearsAtCompany'].mean()

print(f"Total Employees: {total_employees:,}")
print(f"Attrition Rate: {attrition_rate:.2f}%")
print(f"Retention Rate: {retention_rate:.2f}%")
print(f"Average Salary: ${avg_salary:,.0f}")
print(f"Average Experience: {avg_experience:.1f} years")
```

```
attrition_count = (df['Attrition'] == 'Yes').sum()
active_employees = (df['Attrition'] == 'No').sum()

print(f"Employees Left: {attrition_count:,}")
print(f"Active Employees: {active_employees:,}")
```
